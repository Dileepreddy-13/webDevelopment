# Q1: Calculate mean tip amount using numpy (BL-3)
**Aim:** To find the average tip amount given by customers using NumPy, which helps understand the general tipping behavior.

**Syntax:** np.mean(df['column_name']) or df['column_name'].mean()

**Functions:** .mean()

**Logic:** We use df_tips['tip'].mean() to calculate the mean of the “tip” column. NumPy’s mean() function performs the average.

```python
mean_tip = df_tips['tip'].mean() #calculating the mean tip using .mean()
print("Mean Tip Amount:", mean_tip)
```

**Observation:** The mean tip amount is approximately $2.99.

This shows that on average, customers give around $3 as a tip

# Q2: Select entries where party size is greater than 3 in pandas (BL-3)
**Aim:** To filter and view only those records where the dining party had more than 3 members, to analyze tipping patterns for larger groups.

**Syntax:** df[conditon]

**Logic:** We apply a condition df_tips['size']>3 as an index for df_tips.

```python
party_size = df_tips[df_tips['size'] > 3] #selecting entries using boolean indexing
print("Entries where party size is greater than 3 is\n:",party_size)
```

**Observation:** The filtered result shows all large groups (size 4, 5, or 6).These groups are generally having high total bill and tips.

#Q3: Detect and report missing data (BL-4)
**Aim:** To check and report if any data values are missing in the dataset.

**Syntax:** df.isnull().sum()

**Functions:** .isnull(), .sum()

**Logic:** isnull() detects missing values, and sum() counts how many missing entries exist per column.

```python
missing_data = df_tips.isnull().sum() #finding whether a particular data is null or not using .isnull() and counting them using .sum()
print("Number of missing data column wise\n",missing_data)
```

**Observation:** All columns show 0 missing values, meaning the dataset is clean and complete.

No cleaning steps are required before analysis.

#Q4: Group by date and time and calculate average tip (BL-4)
**Aim:** To determine which day and time of the week has the highest or lowest average tip amount.

**Syntax**: df.groupby(['col1','col2'])['column_name'].mean()

**Functions:** .groupby(), .mean()

**Logic:** The groupby() function divides data into groups, and mean() computes the average tip for each group.

```python
avg_tip = df_tips.groupby(['day', 'time'])['tip'].mean() #grouping the data using .groupby() and calculating the average tip
print("Average tip w.r.t date and time\n",avg_tip)
```

**Observation:** This analysis shows that the tipping behaviour is slightly more during week ends than week days

# Q5: Create scatter plot of total bill vs tip colored by day with seaborn (BL-5)
**Aim:** To visualize the relationship between total bill and tip amount and observe how it varies across different days using Seaborn.

**Syntax:**

sns.set_style(style)

plt.figure(figsize=(width, height))

sns.scatterplot(data=<data>, x=<x_col>, y=<y_col>, hue=<category_col>)

plt.title(title_string)

plt.show()

**Functions:**

.set_style()

.figure()

.scatterplot()

.title()

.show()



**Logic:** x and y specify the axes, hue='day' colors the points by day.

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='day')
plt.title("Total Bill vs Tip (Seaborn)")
plt.show()
```

**Observation:** This scatter plot shows that the value of total bill and tips is slightly higher during week ends indicating greater restaurant activity.
